# Pie Chart Representation

```python
 import matplotlib.pyplot as plt 
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

# colour function

```python
 import matplotlib.pyplot as plt 
import numpy as np
y=np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co)
plt.show()
```

# label function

```python
import matplotlib.pyplot as plt 
import numpy as np
y=np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.show()
        
```

# legend functon


```python
import matplotlib.pyplot as plt 
import numpy as np
y=np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.legend()
plt.show()
        
```

```python
import matplotlib.pyplot as plt 
import numpy as np
y=np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.legend(title="subjects")
plt.show()
        
```

# Bargraphs


```python
import matplotlib.pyplot as plt
country=['USA','INDIA','GERMANY','UK','FRANCE']
gdp_per_capita=(48000,38000,50000,45000,44000)
plt.bar(country,gdp_per_capita)
plt.title("BAR GRAPH")
plt.xlabel("country")
plt.ylabel("capita")
plt.show()
```

```python
import matplotlib.pyplot as plt
country=['USA','INDIA','GERMANY','UK','FRANCE']
gdp_per_capita=(48000,38000,50000,45000,44000)
plt.barh(country,gdp_per_capita)
plt.title("BAR GRAPH")
plt.xlabel("country")
plt.ylabel("capita")
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
numbers=[0.1,0.5,1.5,2,4,5,5.5,6,8,9]
plt.hist(numbers)
plt.xlabel("numbers")
plt.ylabel("frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
numbers=[0.1,0.5,1.5,3,4,5]
plt.hist(numbers,bins=6)
plt.xlabel("numbers")
plt.ylabel("frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
numbers=[0.1,0.5,1.5,2,4,5,5.5,6,8,9]
plt.hist(numbers,bins=3,color="green")
plt.xlabel("numbers")
plt.ylabel("frequency")
plt.show()
```
